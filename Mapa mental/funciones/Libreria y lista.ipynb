```python
notasemillero=pd.DataFrame({"Nombres":["pedro", "darden", "leidy", "luis", "marta","andrea", "carlos","felipe","andres", "julian"],
                            "Semestre":[10,9,7,8,10,9,10,8,9,8]})
print(notasemillero)
```

```python
notasemillero["Nota1"]=3
notasemillero
```

```python
notasemillero["Nota2"]=[3.1,4.2,3.3,4.5,4,4,3.8,5,4,3.1]
notasemillero
```

```python
notasemillero.insert(4,"Ciudad",["Manizales","Pereira","Armenia","Cordoba","Popayan","Manizales","Pereira","Manizales","Medellin","Pereira"])
notasemillero
```

Metodo appen, con este puedo omitir el indice ignore_intex=True

```python
notasemillero=notasemillero.append({'Nombres':'Laurita','Semestre':9,'Nota1':3.8,
                                    'Nota2':4,'Ciudad':'Armenia','Departamento':'Quindio'},ignore_index=True)
notasemillero
```
