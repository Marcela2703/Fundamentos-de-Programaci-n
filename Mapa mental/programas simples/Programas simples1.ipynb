Ejercicio 1.

Un camión circula por una carretea a 20m/s . En 5 s , su velocidad pasa a ser de 25 m/s ¿cuál ha sido su aceleración?

```python
# Declaración de variables
var_ent_float_velInicial=0.0
var_ent_float_velFinal=0.0
var_ent_float_tiempo=0.0
var_ent_float_aceleracion=0.0 # Variable que almacena el resultado 
```

```python
# Captura de datos
var_ent_float_velInicial= float(input("Velocidad inicial:"))
var_ent_float_velFinal= float(input("Velocidad final:"))
var_ent_float_tiempo= float(input("Tiempo:"))
```

```python
# Proceso
var_ent_float_aceleracion= (var_ent_float_velFinal - var_ent_float_velInicial) / var_ent_float_tiempo
```

```python
# Salida
print("La aceleración fue:", var_ent_float_aceleracion)
```
