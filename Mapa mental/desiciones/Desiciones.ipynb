**Ejercicio 3:**

Construir una función que reciba como parámetro un entero y retorne la cantidad de digitos.

```python
# Desarrollo de la función
def countDigit(n):
  count = 0
  while n != 0:
      n //= 10
      count += 1
  return count

# Drive Code
n = int(input("Digite el Número: "))
print("Número de Digitos: % d" % (countDigit(n)))              
```

**Ejercicio 4:**

Construir una función que recibira como parámetro un entero y retorne la cantidad de dígitos pares.

```python
# Desarrollo de la función

def countDigit(n):
  Ultdig = str(n)
  count = 0
  for i in range(len(Ultdig)):
    if int(Ultdig[i]) % 2 == 0:
      count = count + 1
  return count


#Llamado de la función
n = int(input("Digite Numero: "))
print("Numero de digitos : % d " % (countDigit(n)))
```

**Ejercicio 9:**

Construir una función que reciba un entero y le calcule su factorial sabiendo que el factorial de un número es el resultado de multipmlicar sucesivamente todos los enteros comprendidos entre 1 y el número dado. El factorial de 0 es 1. Aunque no están definidos los factoriales de números negativos, esta condición se evalúa por fuera de la función antes de su invocación. 

```python
def factorial(n):
    """Función que calcula el factorial de un número entero positivo.
    Parámetros
    n: Es un entero positivo.
    Devuelve el factorial de n.
    """
    tmp = 1
    for i in range(n):
        tmp *= i+1
    return tmp

print(factorial (2) )
print(factorial (4))
```
